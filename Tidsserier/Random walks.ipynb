```python
import numpy as np
import matplotlib.pyplot as plt
```

Vil plotte random walk. Vet egentlig ikek hvorfor det er s√• interessant, men tror det brukes mye i modellering. Akkumulerer tilfeldige avvik. Kan kombinere dette med drift (som er systematisk del av endring over tid)

```python
N = 10
K = 5
x = np.random.normal(size=(N, 2, K))
x = np.r_[np.zeros((1,2,K)), x]
x_cumsum = np.cumsum(x,axis=0)
```

```python
fig = plt.figure(figsize=(8,7))
grid = plt.GridSpec(2, 2)
main_ax = fig.add_subplot(grid[0,:]) 
sub_ax_0 =fig.add_subplot(grid[1,0])
sub_ax_1 =fig.add_subplot(grid[1,1])


main_ax.set(xlabel='x',ylabel='y',
            title='Random walk med uavhengig standardnormalfordelt variabel')
for k in range(K):
    main_ax.plot(x_cumsum[:,0,k],x_cumsum[:,1,k])

sub_ax_0.hlines(0,0,N, linestyle='dashed', color='black')    
sub_ax_0.set(xlabel='t',ylabel='x')
for k in range(K):
    sub_ax_0.plot(x[:,0,k])

sub_ax_1.hlines(0,0,N, linestyle='dashed', color='black')    
sub_ax_1.set(xlabel='t',ylabel='y')
for k in range(K):
    sub_ax_1.plot(x[:,1,k])
    
fig.savefig('random_walk.pdf')
```
