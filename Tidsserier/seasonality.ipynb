```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

from statsmodels.tsa.arima_process import arma_generate_sample
from statsmodels.tsa.seasonal import STL
```

$y=f(t) + u_t$, der $f(t) = t + cos(t)$ og $u_t \sim AR(1)..$

```python
T = 100
```

```python
def generate_ts(slope, σ=2, p=0.5,T=T):
    t = np.arange(T)
    
    trend = slope*t
    season = σ*np.cos((2/12)*t*np.pi)
    u = arma_generate_sample(ar=[1,-p], ma=[1], nsample=T)
    
    y = trend + season + u
    return t,trend, season, u, y
    
```

```python
t, trend, season, resid, y = generate_ts(slope=0.1)
t = pd.date_range(start='jan 1 2010', periods=T,freq='M')
```

```python
fig, ax = plt.subplots()
ax.plot(t, trend, alpha=1)
ax.plot(t, trend+season, c='tab:blue', label='trend+season')
ax.plot(t, y, c='black', alpha=.8, label='realisert')
ax.legend()
```

Vil estimere dekomponeringen fra den realiserte tidserien.. 

veldig greit å estimere trenden med lineær regresjon, men hva med sesongvariasjon? kunne brukt cosinus-funksjonen, men er ikke-lineær.. 

Siden jeg kjenner ground truth kunne jeg dekomponert de greiene eksakt, men i praksis så observerer jeg bare den realiserte tidsserien

```python
from statsmodels.tsa.seasonal import seasonal_decompose, STL
```

```python
data = pd.Series(index=t, data = y)
```

```python
decomposition = seasonal_decompose(data)
```

```python
trend_ = decomposition.trend
seasonal_ = decomposition.seasonal
resid_ = decomposition.resid
```

```python
fig, axes = plt.subplots(2,2, figsize=(10,6), sharex=True)
axes = axes.flatten()

axes[0].plot(t, trend, alpha=1)
axes[0].plot(t, trend+season, c='tab:blue', label='trend+season')
axes[0].plot(t, y, c='black', alpha=.8, label='realisert')

axes[1].plot(t,trend)
axes[1].plot(t,trend_, label='estimert trend', color='black')

axes[2].plot(t, season)
axes[2].plot(t, seasonal_, label='estimert sesongvariasjon',color='black')

axes[3].plot(t, trend_)
axes[3].plot(t, trend_+seasonal_, label='estimert trend+sesong', c='tab:blue')
axes[3].plot(t, y, label='realisert', color='black')

subtitles = ['A.','B.', 'C.', 'D.']

for i in range(len(axes)):
    axes[i].legend()
    axes[i].title.set_text(subtitles[i])
fig.suptitle('Naiv sesongdekomponering')
fig.tight_layout()
```

```python
decomposition = STL(data).fit()
trend_ = decomposition.trend
seasonal_ = decomposition.seasonal
resid_ = decomposition.resid
```

```python
fig, axes = plt.subplots(2,2, figsize=(10,6), sharex=True)
axes = axes.flatten()

axes[0].plot(t, trend, alpha=1)
axes[0].plot(t, trend+season, c='tab:blue', label='trend+season')
axes[0].plot(t, y, c='black', alpha=.8, label='realisert')

axes[1].plot(t,trend)
axes[1].plot(t,trend_, label='estimert trend', color='black')

axes[2].plot(t, season)
axes[2].plot(t, seasonal_, label='estimert sesongvariasjon',color='black')

axes[3].plot(t, trend_)
axes[3].plot(t, trend_+seasonal_, label='estimert trend+sesong', c='tab:blue')
axes[3].plot(t, y, label='realisert', color='black')

for i in range(len(axes)):
    axes[i].legend()
    axes[i].title.set_text(subtitles[i])
fig.suptitle('Sesongdekomponering med lokal regresjon')
fig.tight_layout()
```
