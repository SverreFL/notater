```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x: np.log(x)+np.cos(2*x)
g = lambda x: (1/x)-2*np.sin(2*x)
h = lambda x: -(1/x**2)-4*np.cos(2*x)
```

```python
x_grid = np.linspace(0.3,0.6)
a = 0.5
```

```python
p1 = lambda x:f(a)+g(a)*(x-a)
p2 = lambda x:f(a)+g(a)*(x-a)+0.5*h(a)*(x-a)**2

M = max(np.abs(h(x_grid)))

u = lambda x: f(a)+g(a)*(x-a)+0.5*M*(x-a)**2
l = lambda x: f(a)+g(a)*(x-a)-0.5*M*(x-a)**2
```

```python
fig, (ax0,ax1) = plt.subplots(1,2,figsize=(10,4))

ax0.plot(x_grid,f(x_grid))
ax0.plot(x_grid,p1(x_grid), label='Første orden')
ax0.plot(x_grid,p2(x_grid), label='Andre orden')
ax0.plot(a,f(a),'ro')

ax0.spines['right'].set_visible(False)
ax0.spines['top'].set_visible(False)
ax0.grid()
ax0.legend()

ax1.plot(x_grid,f(x_grid))
ax1.plot(x_grid,p1(x_grid))
ax1.plot(x_grid,u(x_grid),'--',c='grey')
ax1.plot(x_grid,l(x_grid),'--',c='grey')
ax1.plot(a,f(a),'ro')

ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
plt.grid()
ax1.set_xlim(.4,.6)
ax1.set_ylim(-.3,.0)
ax1.set_xticks([0.4,.45,.5,.55,.6])
```

```python
plt.plot(x_grid,f(x_grid))
plt.plot(x_grid,p1(x_grid), label='Første orden')
plt.plot(x_grid,p2(x_grid), label='Andre orden')
plt.plot(a,f(a),'ro')

ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.grid()
plt.legend()
```
