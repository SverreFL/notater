Vil bli mer komfortabel med integrasjon og derivasjoner på $f:\mathbb{R}^2\to\mathbb{R}$

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import numpy as np
from sympy import *
```

```python
def func(x1,x2):
    return x1**2+x1*x2+x2**2
```

```python
xx1, xx2 = np.mgrid[0:1:0.01,0:1:0.01]
yy = func(xx1,xx2)
```

Har en funksjon som er definert på domenet $\{(x_1,x_2):x_i \in [0,1], i=1,2 \}$

```python
fig = plt.figure(figsize=(12,8));
ax = fig.add_subplot(111,projection="3d")

ax.plot_surface(xx1,xx2,yy, alpha=0.5)
ax.set(xlabel='$x_1$',ylabel='$x_2$',zlabel='$f(x_1,x_2)$')
ax.view_init()
```
