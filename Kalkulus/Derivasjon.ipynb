```python
from sympy import *

import matplotlib.pyplot as plt
import numpy as np
```

Vil visualisere derivasjon. Tenker det er et poeng at vi definerer en ny transformasjon $h$ som mapper $s:=\Delta x$ til $\mathbb{R}$. Har en annen inputmengde enn $f$ men vil plotte begge i samme diagram.. slik at jeg får tangentlinje. Det går helt fint dersom vi observerer at $s:=x-a$

tenker jeg bruker lambdify for visualisering ---> få inn i standard framework ved behov nomsayign.

```python
x = symbols('x')
f = 3 + x**3
g = diff(f,x)

grid = np.linspace(0,2)
lam_f = lambdify(x,f)
lam_g = lambdify(x,g)
```

Har nå f og f':=g. Dette er "bare" to funksjoner. Kan bruke dette til å lage tangentlinje. Må da ta utgangspunkt i et gitt element av inputmengden og konstruere en ny funksjon med utgangspunkt i dette.

```python
a = 1
h = g.subs(x,a)*x
h_c = f.subs(x,a)+g.subs(x,a)*(x-a)
lam_h = lambdify(x,h)
lam_h_c = lambdify(x,h_c)
```

```python
fig, axes = plt.subplots(1,3, figsize=(12,4))
axes = axes.flatten()
axes[0].plot(grid,lam_f(grid),label=f'f(x)={str(f)}')
axes[0].plot(grid,lam_g(grid), label=r"$f'(x)$")
axes[0].legend()
axes[0].grid()
axes[0].set(xlabel='x')

axes[1].plot(grid,lam_f(grid),label=f'f(x)={str(f)}')
axes[1].plot(grid,lam_h_c(grid),label=r"$f(1)+f'(1)*(x-1)$")
axes[1].legend()
axes[1].grid()
axes[1].set(xlabel='x')

axes[2].plot(grid,lam_h(grid),label=r"$h(s)=f'(1)s$")
axes[2].legend()
axes[2].grid()
axes[2].set(xlabel='s:=x-a')
```
