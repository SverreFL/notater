```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Mye av intuisjonen min om deriverte er fra funksjoner $\mathbb{R}\to\mathbb{R}$ som kan representeres grafisk med outputmengden perpendikulær på inputmengden i et koordinatsystem.

Mer generelt er en funksjon en mapping av elementer i en mengde til elementer i en annen mengde. 

Kontinuerlig hvis nabolag i inputmengden også tilhører nabolag i outputmengden; blir ikke sendt til ulike steder.

Kan tolke transformasjon som at tallinjen blir strukket, krympet og/eller speilet.

Derivert gir informasjon om transformasjonen lokalt... sier ikke hvor i outputmengden vi er, men sier noe om relasjon mellom nabolag i input og nabolag i output for et gitt punkt i inputmengde.

```python
def plot_derivative(func, ax, x0=0, x1=1, step=.05, a=.2):
    sns.despine(left=True, bottom=True)
    ax.set(xticks=[], yticks=[]);

    grid = np.arange(x0, x1+step, step)
    y_min = min(func(grid))
    y_max = max(func(grid))

    ax.hlines(a, y_min, y_max)
    ax.hlines(1-a, y_min, y_max)
    annotate_text(ax, a=a,y_min=y_min, y_max = y_max)
    
    ax.plot(grid, [1-a]*len(grid), 'ok')
    ax.plot(func(grid), [a]*len(grid), 'ok')
    ax.set_ylim(0,1)
    return ax
    
    
    
```

```python
def annotate_text(ax, a, y_min, y_max):
    ax.text(y_min, 1-a+.05, round(y_min), fontsize=14)
    ax.text(y_max, 1-a+.05, round(y_max), fontsize=14)
    ax.text(y_min, a+.05, round(y_min), fontsize=14)
    ax.text(y_max, a+.05, round(y_max), fontsize=14)
    return ax
```

```python
fig, (ax0,ax1) = plt.subplots(1,2, figsize=(10,5))
plot_derivative(lambda x:2*x, ax0)
ax0.title.set_text('$f(x)=2x$')

plot_derivative(lambda x:2*x**2, ax1)
ax1.title.set_text('$g(x)=2x^2$')
fig.savefig('visual_derivasjon.pdf')
```

Vil utvide til å se på derivasjon av transformasjon $\mathbb{R}^2\to\mathbb{R}^2$. Tenker i utgangspunktet til å se på kvadrat av punkter... eller kanskje vil jeg visualisere den deriverte litt mer direkte.. med piler

lengde på pil er derivert i hver retning.. 

eks: $f(x,y) = x^2+y^2$..

```python
X, Y = np.mgrid[0:1:.1,0:1:.1]
xx, yy = np.mgrid[0:1:.01,0:1:.01]

U = 2*X
V = 2*Y
plt.quiver(X,Y,U,V)
plt.contour(xx,yy,xx**2+yy**2)
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111,projection="3d")
ax.plot_surface(xx,yy,xx**2+yy**2)
```
