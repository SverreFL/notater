```python
N = 50
θ = np.array([1,2])

X = np.c_[np.ones(N), np.random.randn(N)]
y = (X@θ+np.random.randn(N))[:,None]
```

```python
eta = 0.1
n_iter = 10
```

```python
theta = np.random.randn(2,1)

theta_list = []
for iteration in range(n_iter):
    theta_list.append(theta)
    gradient = (2/N)*(X.T@X@theta-X.T@y)
    theta = theta-eta*gradient
```

```python
fig, axes = plt.subplots(2,1, figsize=(8,10))
axes = axes.flatten()
x_grid = np.linspace(min(X[:,1]), max(X[:,1]))

for theta in theta_list:
    axes[0].scatter(theta[0],theta[1])
    axes[1].plot(x_grid, theta[0]+theta[1]*x_grid)
axes[1].scatter(X[:,1],y, c='tab:blue')
```
