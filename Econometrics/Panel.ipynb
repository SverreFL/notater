```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.formula.api as smf

plt.rcParams["font.family"] = "serif" # alltid bytt font
np.random.seed(123)
```

```python
N = 6
u = np.random.normal(size=N)

df = pd.DataFrame(columns=['y','x','d'])
df.x = [1,2,3]*2
df.d = [0]*3+[1]*3
df.y = -4*df.d+df.x+u

df1 = pd.DataFrame(columns=['y','x','d'])
df1.x = [1,2,3,4,5,6]
df1.d = [0]*3+[1]*3
df1.y = -4*df1.d+df1.x+u
df1
```

Viser separat regresjon innad i hver gruppe (egen slope og interecpt, kan også få det med full interaksjon) og pooled OLS som ikke tar hensyn til at de er forskjellig gruppe.

Har samme slope, men ulik intercept (gruppe fasteffekt). Pool gir konsistent estimator dersom ikke korrelasjon mellom gruppetilhørighet og eksponering for behandling. Kan nok være greit å ta med gruppe uansett siden de reduserer størrelse på feilledd.

#### Vil forsøke å vise dekomponering av feilledd

u_it = a_i + e_it

```python
fig = plt.figure(figsize=(10, 10))
grid = plt.GridSpec(2, 2, hspace=0.2, wspace=0.2)
ax0 = fig.add_subplot(grid[0, 0])
ax1 = fig.add_subplot(grid[0, 1])
ax2 = fig.add_subplot(grid[1, :])

colors = ['tab:blue', 'tab:orange']

# plot A
x_grid = np.linspace(min(df.x),max(df.x))
b = smf.ols('y~x',data=df).fit().params
ax0.plot(x_grid,b[0]+b[1]*x_grid,color='black')
for i in (0,1):
    df_sub = df[df.d==i]
    ax0.scatter(df_sub.x,df_sub.y,label=f'i={i}',color=colors[i])
    b = smf.ols('y~x',data=df_sub).fit().params
    ax0.plot(x_grid,b[0]+b[1]*x_grid,color=colors[i],linestyle='dashed')
    ax0.title.set_text('A.')

# plot B
x_grid = np.linspace(min(df1.x),max(df1.x))
b = smf.ols('y~x',data=df1).fit().params
ax1.plot(x_grid,b[0]+b[1]*x_grid,color='black')
for i in (0,1):
    df_sub = df1[df1.d==i]
    ax1.scatter(df_sub.x,df_sub.y,label=f'i={i}',color=colors[i])
    b = smf.ols('y~x',data=df_sub).fit().params
    ax1.plot(x_grid,b[0]+b[1]*x_grid,color=colors[i],linestyle='dashed')
    ax1.title.set_text('B.')

for ax in (ax0,ax1):
    ax.legend()
    ax.grid()
    ax.set_ylim(min(df.y)-.5,max(df.y)+.5)
    ax.set(xlabel='x',ylabel='y')
    

ax2.plot(x_grid,x_grid,color='black',label=r'$\beta\cdot x$')
for i in (0,1):
    df_sub = df1[df1.d==i]
    ax2.scatter(df_sub.x,df_sub.y,color=colors[i])

idx=0    
x = df1.loc[idx,'x']
ymin = df1.loc[idx,'x']
ymax = ymin - 4*df1.loc[idx,'d']
ax2.vlines(x+.05,ymin,ymax,color='tab:red', label=r'$\alpha_i$',alpha=.7)
ymin = ymax
ymax = ymin + u[idx]
ax2.vlines(x,ymin,ymax,linestyles='dotted', label=r'$u_{it}$')
ax2.legend()
ax2.grid()
ax2.title.set_text('C.')
for idx in df1.index[1:]:
    # plotte fasteffekt
    x = df1.loc[idx,'x']
    ymin = df1.loc[idx,'x']
    ymax = ymin - 4*df1.loc[idx,'d']
    ax2.vlines(x,ymin,ymax,color='tab:red',alpha=.7)
    ymin = ymax
    ymax = ymin + u[idx]
    ax2.vlines(x,ymin,ymax,linestyles='dotted')
    

fig.savefig('panel0.pdf')
```
