```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.linalg import inv
```

Vil se på kontrollering for tidstrend.

```python
T = 10
t = np.arange(T)
x = t+np.random.normal(size=T)
y = t + x + np.random.normal(size=T)
```

```python
X = np.c_[np.ones(T),x]
b = inv(X.T@X)@X.T@y
y_hat = X@b
```

```python
plt.plot(t,x,'-o')
plt.plot(t,y,'-o')
plt.plot(t,y_hat,'--',color='tab:orange')
```
