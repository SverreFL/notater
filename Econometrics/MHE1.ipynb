```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

Vil analysere randomisert eksperiment og selection on observables (randomisering innenfor stratifisert sub-samples). Vil helst gjøre det på en måte som kan visualiseres. Føler det er problem at det er vanskelig å visualisere siden blir liksom bare én dimensjon... figurer må liksom nødvendigvis være 2d.

Det er poeng at folk er forskjellig. Tror jeg kan forenkle dette ved å dele de inn etter diskret typer.

```python
N = 10000
x = np.r_[np.zeros(int(N/2)),np.ones(int(N/2))]
D = (x+np.random.uniform(-5,5,size=N) < -1).astype(int)
u = np.random.normal(size=N)
y = 3*x+D+u
df = pd.DataFrame(data=np.c_[x,D,y],columns=['x','D','y'])
```

```python
fig, axes = plt.subplots(3,1,figsize=(6,12))
axes = axes.flatten()
sns.boxplot(x='D',y='y',data=df,ax=axes[0])
sns.barplot(x="x", y="D",ci=None, data=df,ax=axes[1],color='tab:purple')
sns.boxplot(x='x',y='y',hue='D',data=df,ax=axes[2])
titles = ['A.','B.','C.']
for i in range(len(axes)):
    axes[i].title.set_text(titles[i])
    axes[i].xaxis.labelpad = -10
fig.savefig('CIA.pdf')
```

Er poeng at når vi har data så har vi alltid fordeling. Det er bare spørsmål om å se på relativ frekvens.

#### Har lyst til å plotte dårlig kontroll

```python
N = 10000
D = np.random.binomial(1,0.5,size=N)
T = (2*D+np.random.uniform(-5,5,size=N)> 3).astype(int)
u = np.random.normal(size=N)
y = 4*T+0.5*T*D+u
df = pd.DataFrame(data=np.c_[x,D,T,y],columns=['x','D','T','y'])
```

```python
df_sub = df[df['T']>0]
fig, axes = plt.subplots(3,1,figsize=(6,12))
axes = axes.flatten()

sns.boxplot(x='D',y='y',data=df,ax=axes[0])
sns.boxplot(x='D',y='y',data=df_sub,ax=axes[1])
```
