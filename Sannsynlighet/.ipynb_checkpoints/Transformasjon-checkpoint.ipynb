```python
from sympy import *
from sympy.stats import *
import matplotlib.pyplot as plt
import numpy as np
```

Ting har lært:

vil lage FiniteRV ved density() på transformasjon.. får FiniteDensity. 

med FiniteDensity ... dict som mapper value to pmf(value).

bruker som density input i FiniteRV

density() på denne gir Lambda... siden homemade.

bruker lambda på piecewise for å evaluere iterativt inne i sympy

```python
x = Symbol('x', integer=True, positive=True)
n, p = 3, 1/2

X = Binomial('X', n,p)
X_pmf = density(X).pmf(x)
X_pmf = X_pmf.args[0][0] # må huske range ..
```

```python
y = Symbol('y', integer=True, positive=True)
g = Lambda(x,2*x)
Y = FiniteRV('Y', density(g(X)))
Y_pmf = density(Y).pmf(y)
```

Transformasjon av diskret gir FiniteDensity.. som i praksis er en dict

```python
x_vals = np.arange(n+1)
y_vals = [float(g(x)) for x in x_vals]

x_pmf_vals = [float(X_pmf.subs(x, v)) for v in x_vals]
y_pmf_vals = [float(Y_pmf(y)) for y in y_vals]
```

```python
fig, (ax0, ax1) = plt.subplots(1,2, figsize=(8,4))
fig.suptitle(r'pmf av $Y=2\cdot X$')

ax0.plot(x_vals, x_pmf_vals,'o')
ax0.vlines(x_vals,0, x_pmf_vals)
ax0.set(xlabel='x', ylabel='$p_X(x)$', ylim=(0,.4), xticks=x_vals)

ax1.plot(y_vals, y_pmf_vals,'o')
ax1.vlines(y_vals,0, y_pmf_vals)
ax1.set(xlabel='y', ylabel='$p_Y(y)$', ylim=(0,.4))

fig.tight_layout()
fig.savefig('transformert_pmf.pdf')
```

Vil nå gå over til transformasjon av kontinuerlig fordeling

```python
X = Uniform('X',0,1)
X_dist = density(X)
X_pdf = Lambda(x,X_dist.pdf(x))
```

```python
g = Lambda(x,2*x**2)
Y = g(X)
Y_pdf = density(Y)
```

```python
g_inv = Lambda(y,solve(Eq(g.expr,y),x)[0])
g_inv_dy = Lambda(y,g_inv(y).diff(y))
```

```python
fig, axes = plt.subplots(2,2, figsize=(10,6))
axes = axes.flatten()
x_grid = np.linspace(0,1)
y_grid = np.linspace(0,2)

axes[0].plot(x_grid, [X_pdf(v) for v in x_grid])
axes[0].set(xlabel='x', ylabel='$f_X(x)$', ylim=(0,2))

axes[1].plot(x_grid, [g(v) for v in x_grid], label='$g(x)=2x^2$')
axes[1].set(xlabel='x', ylabel='y')
axes[1].legend()

axes[2].plot(y_grid, [g_inv(v) for v in y_grid], label='$g^{-1}(y)$')
axes[2].plot(y_grid, [g_inv_dy(v) for v in y_grid], label=r'$\frac{d}{dy}g^{-1}(y)$')
axes[2].set(xlabel='y')
axes[2].legend()

axes[3].plot(y_grid, [Y_pdf(v) for v in y_grid])
axes[3].set(xlabel='y', ylabel='$f_Y(y)$',ylim=(0,2))

subtitles = ['A.', 'B.', 'C.', 'D.']
for i in range(4):
    axes[i].title.set_text(subtitles[i])
    
fig.tight_layout()
fig.savefig('transformert_pdf.pdf')
```
