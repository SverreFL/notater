Har ofte simultan fordelinger på form

$$f(x,y) = e^{-y}, 0<x<y<\infty$$

altså der selve funksjonsverdien bare avhenger av en variabel, men support avhenger av annen variabel. Vet liksom alltid at den ene må være større enn den andre. Vil forsøke å jobbe litt med denne formen og se om jeg kan visualisere den.

```python
x,y = symbols('x y', positive=True)
f_xy = E**(-y)
```

Vil først sjekke at dette er gyldig simultanfordeling ved å integrere over support.
$$\int_0^{\infty}\int_x^{\infty}f(x,y)dydx$$



```python
inner = integrate(f_xy,(y,x,oo))
integrate(inner,(x,0,oo))
```

Vil nå forsøke å få ut marginale og sjekke at de er gyldige. Kan jo merke at inner tilsvarer marginal av x. (eller???) Så har allerede sett at den er gyldig. Kan jeg plotte denne funksjonen.

```python
%matplotlib inline
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111,projection="3d")

grid = np.linspace(0,5,200)
xs,ys = np.meshgrid(grid,grid)
zs = np.exp(-ys)

domain = ys<xs
zs[domain==False] = 0

ax.plot_surface(xs,ys,zs,alpha=.8)
ax.view_init(20,20)
ax.set(xlabel='x',ylabel='y',zlabel='f(x,y)');
```
