```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
N = 50000
x = np.random.choice(np.arange(1,6+1),(2,N))
x = np.cumsum(x,0)
```

```python
unique, counts = np.unique(x[0,:],return_counts=1)
counts = counts/sum(counts)
plt.vlines(unique,0,counts,lw=3) #lw er linewidth
```

```python
fig, axes = plt.subplots(2,3,figsize=(10,6))
axes = axes.flatten()

for i, k in enumerate(unique):
    axes[i].title.set_text(f'f(y|X={k})')
    
    subset = x[1,:][np.where(x[0,:] == k)]
    unique, counts = np.unique(subset,return_counts=1)
    counts = counts/sum(counts)
    axes[i].vlines(unique,0,counts,lw=3)
    axes[i].set_xticks(unique)
```

```python
fig, axes = plt.subplots(3,4,figsize=(10,10))
axes = axes.flatten()

for i, k in enumerate(np.arange(2,12+1)):
    axes[i].title.set_text(f'f(x|Y={k})')
    
    subset = x[0,:][np.where(x[1,:] == k)]
    unique, counts = np.unique(subset,return_counts=1)
    counts = counts/sum(counts)
    axes[i].vlines(unique,0,counts,lw=3)
```
