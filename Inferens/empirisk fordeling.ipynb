```python
import numpy as np
from scipy.stats import norm
from matplotlib import pyplot as plt
import seaborn as sns
from matplotlib import rc


sns.set_style("ticks", {"xtick.major.size": 4, "ytick.major.size": 4})
```

Har et utvalg med realiserte observerasjon.. for hver verdi i grid vil jeg finne andel av observasjon i utvalg som er mindre enn gitt verdi.

```python
def ecdf(x,s):
    return (x <= s).mean()
```

```python
q = norm()

grid = np.linspace(-3.5, 3.5, num=200)

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))
axes = axes.flatten()
N = 10, 25, 50, 250
for i in range(4):
    ax = axes[i]
    n = N[i]
    with sns.color_palette("muted"):
        ax.plot(grid, q.cdf(grid), label=r"$F$")
        for j in range(10):
            obs = q.rvs(n)
            ax.plot(grid, [ecdf(obs,s) for s in grid])
        ax.legend(loc=2)
        ax.set(xlim=[-3.5,3.5], ylim=[0,1],xticks=[-3.5,3.5],yticks=[0,1])
        ax.title.set_text(f'N = {n}') 
fig.savefig('ecdf.pdf')        
```

```python
def ecdf_k(x, h=h, obs=obs):
    N = len(obs)
    return (1/N)*sum(q.cdf((x-obs[n])/h) for n in range(N))
```

```python
obs = q.rvs(25)
```

```python
fig, ax = plt.subplots(figsize=(8,5))
sns.rugplot(x=obs, height=0.05, color='black', ax=ax)
ax.plot(grid, q.cdf(grid), color='black', label='cdf')
sns.ecdfplot(obs, ax=ax, label='ecdf')

for h in (0.1,.5,1, 1.5):
    ax.plot(grid, ecdf_k(grid, h, obs=obs), label=f'h={h}')
ax.legend()
ax.set(ylim=(0,1),title='ECDF med kernel')
fig.savefig('ecdf_kernel.pdf')
```
