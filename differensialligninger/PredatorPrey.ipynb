Har sett på modellering av enkelt populasjon. Kan beskrive det med eksponentiell vekst
$$
y'=\alpha y
$$
eller med logistisk vekst
$$
y' = \alpha y\left(1-\frac{y}{B}\right)
$$

Vil forsøke å analysere disse først. Deretter vil jeg gå videre til en predator-prey modell der det er to populasjoner. Det gir oss et ligningssystem
\begin{align}
x' &= \alpha x - xy \\
y' &= - \beta y + xy
\end{align}

```python
from sympy import *
import matplotlib.pyplot as plt
import numpy as np
```

Skal prøve å løse logistisk vekst først

```python
t, r, K = symbols('t r K', positive=True)
y = symbols('y', cls=Function)
r = 1/2
K = 10

diff_eq = Eq(y(t).diff(t), r*y(t)*(1-(y(t)/K)))
```

```python
diff_eq
```

```python
classify_ode(diff_eq)
```

```python
sol = dsolve(diff_eq).args[1]
```

```python
sol_particular = sol.subs([(K,10),(r,0.2),('C1',4)])
```

```python
plot(sol_particular,(t,0,10))
```

Litt usikker på valg av C.. kommer ikke så veldig langt med de analytiske greiene. Vil har retningsdiagram til de fuckings kurvene mine

```python
y_diff = solve(diff_eq,y(t).diff(t))[0]
lam_y_diff = lambdify(args=[t,y(t)],expr=y_diff)

# lage grid, går ut i fra at t bør begynne i 0, litt usikker på optimalt valg av grid
tt, yy = np.mgrid[0.1:20:1,0:15:.5]
U = np.ones(shape=tt.shape)
V = lam_y_diff(tt,yy)

# normaliserer lengde på vektor
length = np.sqrt(U**2 + V**2)
U = U/length
V = V/length

# angles='xy' er nødvendig for at blir riktig. vet ikek hvorfor
fig, ax = plt.subplots(figsize=(12,8))
ax.quiver(tt,yy,U,V,headaxislength=0, units='width', angles='xy') 
```

Skal nå utvide til ligningssystem... kunne tegne partikulære løsningskurver $(t,x(t))$ og $(t,y(t))$ i samme diagram... men bedre å lage parametriske løsningskurver i $xy$ diagram

Vil lage predator prey modell der jeg ser bort fra de fuckings parametrene mann

```python
x, y = symbols('x y', cls=Function)
t = symbols('t')
ode_x = Eq(x(t).diff(t), x(t)-(1/2)*x(t)*y(t))
ode_y = Eq(y(t).diff(t), -(1/2)*y(t) +(1/2)*x(t)*y(t) )
```

Ser ut som det er vanskelig å løse analytisk.. uansett, vil forsøke å lage vector field

$\frac{dx}{dt} = ...$

```python
x_t = ode_x.args[1]
y_t = ode_y.args[1]
```

```python
x_t
```

plugger bare inn masse vilkårlige verdier av $x=x(t)$ og $y=y(t)$ uten å bry seg om hverken $t$ eller selve funksjonen.. ting er operativt siden jeg bare trenger output. Kan da finne lokalt hvordan en parametrisk kurve vil bevege lokalt for alle steder i $xy$-diagram

```python
lam_x_t = lambdify([x(t),y(t)], x_t)
lam_y_t = lambdify([x(t),y(t)], y_t)
```

```python
# lage grid, går ut i fra at t bør begynne i 0, litt usikker på optimalt valg av grid
tt, yy = np.mgrid[0:3:.2,0:3:.2]
U = lam_x_t(tt,yy)
V = lam_y_t(tt,yy)

# normaliserer lengde på vektor
length = np.sqrt(U**2 + V**2)
U = U/length
V = V/length

# angles='xy' er nødvendig for at blir riktig. vet ikek hvorfor
fig, ax = plt.subplots(figsize=(12,8))
ax.quiver(tt,yy,U,V,headaxislength=0, units='width', angles='xy') 
```
