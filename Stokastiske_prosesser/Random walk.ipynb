```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

Tror random walks er sum av uavhengige variabler.

```python
N = 1000
M = 6
x = np.random.normal(size=(N,M))
z = np.cumsum(x,axis=0)
```

```python
for m in range(0,M,2):
    plt.plot(z[:,m],z[:,m+1])
```

vil flippe kron/mynt mange ganger.

```python
N = 20
M = 1000
x = np.random.choice([0,1],size=(N,M))
z = np.cumsum(x,axis=0)
```

Skal se hvordan fordeling av antall "treff" (1) avhenger av N. Vil ha binomialfordelinger.

```python
n_vals = (1,5,10,15)
fig, axes = plt.subplots(2,2,figsize=(10,6))
axes = axes.flatten()
for i,n in enumerate(n_vals):
    unique,counts = np.unique(z[n-1,:],return_counts=1)
    counts = counts/sum(counts)
    axes[i].vlines(unique,0,counts,color='tab:blue',lw=4)
    axes[i].set_xticks(unique)
    pmf = stats.binom(n=n,p=.5).pmf(unique)
    axes[i].plot(unique,pmf,'o',c='tab:orange',alpha=.5)
```
