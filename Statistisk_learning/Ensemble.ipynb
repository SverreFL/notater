```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)

from matplotlib.colors import ListedColormap
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import make_moons
```

Vi første vise at kombinasjon av svake lærere blir mega sterk lærer.

```python
N = 10000
M = 10
x = np.random.binomial(1,0.51,size=(N,M))
p = np.true_divide(x.cumsum(axis=0),np.arange(1,N+1)[:,None])
```

```python
fig, ax = plt.subplots(figsize=(8,6))
for m in range(M):
    ax.plot(p[:,m], alpha=.8)

ax.plot([0.5]*N, c='black')
ax.plot([0.51]*N, c='black', linestyle='--')
ax.set(ylim=(0.4,0.6), xlabel='Antall learners', ylabel='p', xlim=(0,N))
ax.title.set_text('Sannsynlighet for riktig prediksjon med ensemble av uavhengige learners')
fig.savefig('LLN_ensemble.pdf')
```
