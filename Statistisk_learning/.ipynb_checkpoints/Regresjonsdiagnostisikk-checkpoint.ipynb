```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

Kommer veldig langt med kryssvalidering til å oppnå god fit. Tenker kanskje derfor at det er bedre å flytte dette til økonometri der kryssvalidering ikke løser alle problemer, men på en annen side er jo statistisk læring mer enn prediksjon og det er ikke alltid vi har overflod av data. Noen ganger vil vi ha enkle modeller som er enkle å tolke. Men da må vi også passe på at de ikke blir *for* enkle

```python
N = 100

x = np.random.uniform(-1,1,N)
y = 0.5*x+2*x**2+np.random.normal(1,0.5,size=N)

mod_0 = sm.OLS(y,sm.add_constant(x)).fit()
mod_1 = sm.OLS(y,sm.add_constant(np.c_[x,x**2])).fit()
```

Per konstruksjon ukorrelert med covariates og siden yhat er lineær funksjon av covariates så er ukorrelert der også. Men vi vil ha mer!

```python
fig, axes = plt.subplots(3,1,figsize=(8,12))
axes = axes.flatten()

x_grid = np.linspace(min(x),max(x))

axes[0].scatter(x,y)
axes[0].set(xlabel='x',ylabel='y')
sm.graphics.abline_plot(model_results=mod_0,ax=axes[0],label='model 1',linestyle='dotted')
axes[0].plot(x_grid,mod_1.params[0]+mod_1.params[1]*x_grid+mod_1.params[2]*x_grid**2,
             label='model 2', linestyle='dashed')
axes[0].legend(framealpha=0)

axes[1].scatter(mod_0.fittedvalues,mod_0.resid, edgecolor='tab:blue',facecolor='none')
axes[1].hlines(0,min(mod_0.fittedvalues),max(mod_0.fittedvalues),linestyle='dotted',color='tab:blue')
axes[1].set(xlabel='$\hat{y}$',ylabel='$\hat{u}$')

axes[2].scatter(mod_1.fittedvalues,mod_1.resid, edgecolor='tab:blue',facecolor='none')
axes[2].hlines(0,min(mod_1.fittedvalues),max(mod_1.fittedvalues),linestyle='dashed',color='tab:blue')
axes[2].set(xlabel='$\hat{y}$',ylabel='$\hat{u}$')

titles = ['A.', 'B.', 'C.']
for i in range(3):
    axes[i].title.set_text(titles[i])
    axes[i].title.set_position((0.05,1))
```

```python
fig.savefig('reg_diagnostikk.pdf')
```
