```python
def plot_decision(clf, X, y, ax, scatter=True, treshold=.5):
    x0_min, x0_max = min(X[:, 0])-.2, max(X[:, 0])+.2
    x1_min, x1_max = min(X[:, 1])-.2, max(X[:, 1])+.2
    xx0, xx1 = np.mgrid[x0_min:x0_max:.01, x1_min:x1_max:.01]
    
    clf = clf.fit(X,y)
    try:
        yy = clf.predict_proba(np.c_[xx0.ravel(), xx1.ravel()]).reshape(xx1.shape)
    except:
        yy = clf.decision_function(np.c_[xx0.ravel(), xx1.ravel()]).reshape(xx1.shape)
        print('bruker decision_function')
    
    ax.contourf(xx0, xx1, yy, cmap='RdBu_r')
    ax.contour(xx0,xx1,yy,levels=[treshold], linestyles='dashed', linewidths=3, colors='black')
    if scatter:
        ax.scatter(X[:, 0], X[:, 1], c=y, cmap='RdBu_r')
    
    ax.set(xlim=(x0_min,x0_max), ylim=(x1_min, x1_max))
    ax.title.set_text(str(clf))
#    return ax
```

```python
clf = LogisticRegression()
fig, ax = plt.subplots()
plot_decision(clf, X, y, ax, treshold=-3)
```
