```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import cross_val_score, GridSearchCV, learning_curve
from sklearn.metrics import mean_squared_error, make_scorer
```

Vil først undersøke for utvalg med gitt størrelse, deretter se på hvordan ting avhenger av størrelse på utvalg (varians går ned, kan ha mer fleksibel modell)

```python
N = 50
x_min, x_max = -2, 6
x = np.random.uniform(x_min, x_max, size=N)
y = np.c_[np.cos(x)+np.sin(x)+0.5*np.random.randn(N)]
```

```python
steps = [('scaler',StandardScaler()),
         ('poly', PolynomialFeatures()),
         ('lin_reg', LinearRegression())]
lin_reg = Pipeline(steps).fit(x[:,None],y)
```

```python
x_grid = np.linspace(x_min, x_max, 200)
y_true = np.cos(x_grid)+np.sin(x_grid)

degrees = [2,3,4]

plt.scatter(x,y, edgecolors='tab:red', facecolor='none')
plt.plot(x_grid,y_true,color='tab:red', linestyle='--',label='$\cos(x)+\sin(x)$')

for d in degrees:
    lin_reg = lin_reg.set_params(poly__degree = d).fit(x[:,None],y)
    y_pred = lin_reg.predict(x_grid[:,None])
    plt.plot(x_grid, y_pred,'-', label=d)


    plt.legend(framealpha=0)
```

```python
M = 100
d_grid = np.arange(1,10+1)

for d in d_grid:
    for m in range(M):
        lin_reg = lin_reg.set_params(poly__degree = d).fit(x[:,None],y)
```

Bias og varians er teoretiske egenskaper som avhenger av funksjonen $h$ og simultanfordelingen som generer data. Avhenger ikke av det partikulære utvalget vi observerer, men siden det ble generert fra fordelingen kan vi bruke det til å estimere .. sum av bias^2 + varians + irreducible error. Kan ikke dekomponere dersom ikke kjenner fordeling.

bias^2 + varians for gitt h... må ha noe med utvalget å gjøre. Ikke bare h. Egenskap ved vår estimator for h..... mind fucket som faen her.


```python
N = 50
x_min, x_max = -2, 6
x = np.random.uniform(x_min, x_max, size=N)
y = np.cos(x)+np.sin(x)+0.5*np.random.randn(N)
```

```python
steps = [('poly', PolynomialFeatures()),
         ('lin_reg', LinearRegression())]
lin_reg = Pipeline(steps).fit(x[:,None],y)
```

- Irreducible: $E[(y-f^*(x))^2]$ # teoretisk
- Bias_sq: $E[(f^*(x)-h^*(x))^2]$ # teoretisk
- Varians: $E[(H^*(x)-\hat{h}(x))^2]$, avhenger av N! fordi $\hat{h}$ er estimator fra utvalg # simulering


```python
N_pop = 10000
x_pop = np.random.uniform(x_min, x_max, size=N_pop)
y_pop = np.cos(x_pop)+np.sin(x_pop)+0.5*np.random.randn(N_pop)
irreducible = mean_squared_error(y_pop,np.cos(x_pop)+np.sin(x_pop))

degrees = np.arange(6+1)
bias2 = []
var = []
mse = []

for d in degrees:
    # bruker hele populasjon til å finne beste linje som kalkuere bias
    lin_reg_pop = lin_reg.set_params(poly__degree=d).fit(x_pop[:,None],y_pop[:,None])
    y_best_pred_d = lin_reg_pop.predict(x_pop[:,None])
    bias2.append(mean_squared_error(np.cos(x_pop)+np.sin(x_pop), y_best_pred_d))
    
    # kjører simuleringer av utvalget til å beregne MSE i estimering
    mse_estimation = []
    cv_score = []
    for m in range(100):
        x_sample = np.random.uniform(x_min, x_max, size=N)
        y_sample = np.cos(x_sample)+np.sin(x_sample)+0.5*np.random.randn(N)
        lin_reg_sample = lin_reg.set_params(poly__degree=d).fit(x_sample[:,None],y_sample[:,None])
        mse_estimation.append(mean_squared_error(y_best_pred_d, lin_reg_sample.predict(x_pop[:,None])))
        cv_score.append(cross_val_score(lin_reg_sample,x_sample[:,None],y_sample).mean())
    var.append(np.mean(mse_estimation))
    mse.append(np.mean(cv_score))
```

```python
plt.plot(degrees,bias2,'-o', label='bias2')
plt.plot(degrees, var,'-o', label='var')
plt.plot(degrees, mse, '-o', label='mse')
plt.plot(degrees, [irreducible]*7, label='irreducible')
plt.legend()
```
