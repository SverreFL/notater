##### Colormap

Når vi lager plots er det mange Aritster som inheriter fra ScalarMappable (scatter, imshow, pcolor,..) og derfor har felles argument,
- `cmap`, hvilke fargekart
- `vmin`, hvilken verdi som korresponderer med bunn
- `vmax`, og topp. Kan inferres fra data, men greit å spesifisere dersom bruke samme kart på flere axes
- `norm`, hvis vi for eksempel vil ha logaritmisk skalering i valg av farge

disse argumentene feeded inn i im = ax.[func](..., cmap=..,...)

Deretter bruker vi fig.colorbar(im, ...). Dette er en figurmetode som i default krymper den eksisterende axes og legger til en ny axes inne i figuren der colorbar får plass. Kan eventuelt lage colorbar axes (cax) manuelt å spesifisere det som argument.

Må velge colormap som formidler riktig informasjon til leser av figuren.
1. Sequential: kontinuerlig verdi, lav til høy, gjerne samme farge med ulik intensitet. Blå hvis bra, rød hvis dårlig.
2. Diverging: tilsvarende som over, men med naturlig midtpunkt/nullpunkt. Blå til rød.
3. Kvalitativ: diskret, ulike farger som ikke gir inntrykk av bra eller dårlig.

Finnes lang rekke colormap objekter i matplotlib. For disse er det tilstrekkelig å spesifisere navn som string. Kan i prinsippet lage mine egne ved å spesifisere farger, men tror ikke dette er relevant.

Kan reversere colormap ved å legge til `_r` på slutten, eg: `RdBu_r`

### Generelt om visualisering

Hentet fra [Fundamentals of Data Visualization](https://clauswilke.com/dataviz/).



Vi kan skille mellom to typer datavisualisering

1. Utforskende datavisualisering

Første steg i analysen er å bli litt kjent med data. Vi kan se om det bekrefter våre fordommer og om vi kan oppdage noen mønstre. Videre kan vi identifisere ekstremverdier (uteliggende observasjon) som kan ha stor inflytelse på modeller, samt andre artifacts fra datainnsamlingen (inkludert feilkodinger). 

I dette steget er det viktig å kunne lage raske figurer siden vi trenger mange ulike *views* på data. Det er da en stor fordel med interaktivitet siden vi kan utforske egenskaper til ulike subset av data på rask måte. Kan riktignok komme ganske langt med statiske trellis plot.. 

2. Forklarende datavisualsering

En annen situasjon er nå vi allerede har en konklusjon eller innsikt som vi ønsker å formidle gjennom en figur (grafisk representasjon). Da vil vi gjerne ha litt mer low-level kontroll og ikke like viktig med interaktivitet.

### Grammatikk

Datavisualisering er representasjon av data som formidler informasjon. I likhet med andre medium for formidling bør fremstillingen være ryddig og strukturert. Den bør også være estetisk tiltatelende. Fremstillingen må være korrekt og ikke-misvisende. 

Representasjonen innebærer at det er en mapping mellom data og de grafiske elementene i figuren. Hvert av elementene har ulike egenskaper eller `aesthetics`, i hovedsak
1. Posisjon
2. Form
3. Farge
4. Størrelse

Vi kan betegne mapping mellom data og egenskap som en `scale`. For at figuren skal være ryddig og oversiktelig må mappingen være tydelig og intuitiv slik at leser ikke må bruke energi på å fortolke selve representasjonen. Det er vesentlig om data vi mapper fra er kategorisk, ordinal eller numerisk (?). 
- Hvis kun kategori så er det ingen rangering og representasjonene bør være distinkte og likeverdige. 
- Hvis ordinal må vi ta hensyn til om representasjon formidler rangeringen. 
- Hvis numerisk må vi også ta hensyn til representasjon av avstand mellom verdi. 

#### Posisjon

I 2d figur kan posisjon gi informajon om to variabler gjennom å representere element i kartesisk produkt. Det er er viktig med axis labels og ticks som gjør verdier og måleenhet tydelig for å mappe tilbake fra plassering til underliggende data. 

Det kan være aktuelt å bruke ikke-lineær akse der det ikke er like stor avstand mellom hvert ticks. Kan bruke logaritimisk akse der vi må gange med gitt tall for å gå mellom ticks. Trenger ikke transformere data; representere verdi i fra opprinnelig skala (kunne ekvivalent ha logtransformert data og hatt lineær skala på axis, men dette er mer forvirrende).

#### Farge

Farge har i hovedsak tre bruksområder:
1. Skille mellom distinkte elementer
2. Mappe til verdi
3. Fremheve noe

Valge av fargepalette avhenger av bruk. 

1. I førstnevnte er det bare koding av ulike kategori. Det er da viktig at fargevalget er nøytralt (likeverd) og gjør det mulig å skille. 
2. Hvis verdi er ordinal er det til tilstrekkelig at det gir en rangering. Kan være naturlig å ta fargevalg ut fra om høyere verdi er bra eller dårlig. Bruker to farger kun hvis det har et naturlig midtpunkt. Hvis numerisk må helst avstand i farge korrespondere med avstand i verdi.
3. For å fremheve er det viktig at resten av data (som gir kontekst/kontrast for verdi vi er interessert i) ikke konkurrerer om oppmerksomheten. Bruker `accent` der noen farger er muted..

#### Noen fallgruver

- Proportional ink: Hvis noe er fargelagt, så må areal være proposjonal med verdi det indikerer. Dette medfører at barplots må begynne på 0. Hvis vi fargelegger areal under graf må dette også begynne på 0. Alternativet er ikke teknisk feil, men gir misvisende fremstilling.
- Må håndtere overlappene elementer: Kan bruke opacity/jitter på punkter..
- Ikke bruk for mange farger: Det er vanskelig å skille og holde oversikt. Hvis vi må encode mange distinkte element kan det være bedre å bruke tekst inni figuren. Kan generelt bruke direkte encoding eller legend.
- Ikke gjør for mye samtidig: Bruk mange gjentatte plot på ulike subsets av data. Samme axis gjør det mulig å se på mønster, avvik og trender. Annen måte er å lage sammensatt figur av ulike typer plot med helhetlig stil (fargevalg/fonts mm.).


#### Prinsipp for utforming av tabeller

1. Ikke bruk vertikale linjer
2. Ikke bruk horisontale linjer mellom ulike rekke av samme type
3. Tekst er left-aligned
4. Tall er right-aligned med samme antall desimal
5. Enkeltsymbol er centered
6. Kolonnenavn har samme alignment som innhold

#### Litt om design

Vi bør generelt begrense mengden av visuelle elementer (*ink*) som ikke representer data (eg: ramme på figur, grid i bakgrunnen, tekst). Det gir et mer stilrent og elegant design. Samtidig må det være tilstrekkelig struktur for at vi enkelt kan omkode representasjonen. Hvis det er for minimalistisk kan det gi inntrykk av at elementer flyter rundt i rommet og det er vanskelig å knytte posisjon til underliggende verdi. 

Figur bør ha nok informasjon til å fjerne ambiguitet i representasjon, men ikke inneholde noe overflødig. Lurt å bruke større fontsize enn default, spesielt på axis label.

#### Filformat

##### Bitmap

Representere informasjon om grid av punkter (pixels). Har ulike filformat med litt ulike strategier for dekomprering som reduserer antall bits som er nødvendig for å representere innhold. Hvis vi for eksempel vet at det er en svart strek som går horisontalt så er det ikke nødvendig å representere hver pixel individuelt for å rekonstruere innhold.

- Lossless hvis ikke tap av informasjon
- Lossy hvis noe tap

for fotografi og sånn kommer vi ikke så langt med lossless, men kan kan dekomprere en del uten at det blir synlig for menneskelig øye (hvis ikke zoomer inn..). Medfører såkalte *artifacts*. 

##### Vector

I stedet for å representere innhold ut fra pixler, så representere det individuelle grafiske element og deres relative plassering. Det medfører at vi kan zoome inn uten at figur blir pixelert siden antallet pixler som representere element blir oppdater. Har i hovedsak to format
1. pdf
2. svg (bedre for nettleser-ting tror jeg...)

Nedsiden med vector er at de grafiske elementene må blir renderet av klient (nettleser, adobe reader, mm.). Dette kan ta lang tid hvis veldig mange element (scatter med individuelle element), filstørrelse kan bli stor og det kan oppstå problem dersom klient ikke kjenner til element vi bruker (eks: font).
