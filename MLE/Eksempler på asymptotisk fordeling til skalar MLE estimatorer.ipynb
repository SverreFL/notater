```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats
```

```python
N = 100
M = 1000
σ, λ, p = 4,4,0.5

normal = stats.norm(0,σ).rvs((N,M)).mean(axis=0)
poisson = stats.poisson(λ).rvs((N,M)).mean(axis=0)
bernoulli = stats.bernoulli(p).rvs((N,M)).mean(axis=0)

```

```python
grid = np.linspace(-3,3)

fig, (ax0,ax1,ax2) = plt.subplots(3,1,figsize=(10,10))
ax0.plot(grid,stats.norm(0).pdf(grid))
sns.distplot(np.sqrt(N)*(normal/σ), ax=ax0)

ax1.plot(grid,stats.norm(0).pdf(grid))
sns.distplot(np.sqrt(N)*(poisson-λ)/np.sqrt(λ),ax=ax1)

ax2.plot(grid,stats.norm(0).pdf(grid))
sns.distplot(np.sqrt(N)*(bernoulli-p)/np.sqrt(p*(1-p)),ax=ax2)
```
