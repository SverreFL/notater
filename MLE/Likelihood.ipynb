```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, bernoulli, binom
```

Vil illustrere likelihood. Kan koble til urne modell med tilbakelegging: Hvis to farger så er det bernolli. Relativ andel er p. Hvis vi vet dette kan vi finne sannsynlighet for antall av hver farge vi trekker med binomialfordeling. Men hva hvis gjør noen trekk og heller vi si noe om andel av hver? Dette er likelihood!

```python
x = [1,1,0,1,0,0]
m = 50
p_grid = np.linspace(0,1,m)

l_grid = np.empty((m,len(x)))
for i in range(len(x)):
    l_grid[:,i]=bernoulli(p_grid).pmf(x[i])
l_grid = np.cumprod(l_grid,axis=1)
```

```python
fig, axes = plt.subplots(2,3,figsize=(10,8))
axes = axes.flatten()
for i in range(6):
    ax = axes[i]
    ax.plot(p_grid,l_grid[:,i])
    ax.title.set_text(f'x={x[i]}')
    ax.set_yticks([])
fig.savefig('likelihood_bernoulli.pdf')
```

```python
x = norm().rvs(6)
m = 50
μ_grid = np.linspace(-2,2,m)

l_grid = np.empty((m,len(x)))
for i in range(len(x)):
    l_grid[:,i]=norm(loc=μ_grid).pdf(x[i])
l_grid = np.cumprod(l_grid,axis=1)
```

```python
fig, axes = plt.subplots(2,3,figsize=(10,8))
axes = axes.flatten()
for i in range(6):
    ax = axes[i]
    ax.plot(μ_grid,l_grid[:,i])
    ax.title.set_text(f'x={x[i]:.2f}')
    ax.set_yticks([])
fig.savefig('likelihood_norm.pdf')
```
