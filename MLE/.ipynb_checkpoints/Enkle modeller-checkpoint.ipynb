```python
from sympy import *
import sympy.stats as st
from scipy import stats

import matplotlib.pyplot as plt
import numpy as np
```

### Normal med kjent varians

```python
x, θ = symbols('x θ')
dists = [st.Normal('X',θ,1), st.Cauchy('X',θ,1)]
```

```python
N = 10
grid = np.linspace(2,6)
fig, axes = plt.subplots(2,2,figsize=(10,7))

for i in (0,1):
    d = dists[i]
    L = st.density(d).pdf(x)
    logL = log(L,θ)
    score = diff(logL,θ)
    
    for n in range(N):
        obs = st.sample(d.subs(θ,4))
        logL_obs = logL.subs(x,obs)
        score_obs = score.subs(x,obs)
        
        logL_vals = lambdify(θ,logL_obs)(grid)
        score_vals = lambdify(θ,score_obs)(grid)
        axes[0,i].plot(grid,logL_vals)
        axes[1,i].plot(grid,score_vals)
        
        
    
    
```

```python
class Likelihood():
    def __init__(self,dist,θ,continious=True):
        '''L: likelihoodfunksjon som expression'''
        self.dist = dist
        self.θ = θ
        x = symbols('x',real=True)
        if continious:
            self.L = st.density(dist).pdf(x)
        else:
            self.L = st.density(dist).pmf(x)
        self.logL = log(self.L)
        self.score = diff(self.logL,θ)
        self.hessian = diff(self.score,θ)
        self.fisher = -1/self.hessian
    
    def mle_sample(self,N):
        logL = sum([self.logL for n in range(N)]) 
        score = diff(logL,θ)
        θ_hat = solve(score,θ)[0]
        return θ_hat
```
